#### Group By in mongoDB using Pipeline

```python
pipeline = [
    {
    "$group": {
        "_id": "$Country",
        "Num_Of_Customers" : {"$sum": 1}
        }
    },
    
    {
    "$sort":{
        "Num_Of_Customers": -1
        }
    }
]

list(db.Customers.aggregate(pipeline))
```
